## Swimlane Tools Examples:



```python
import networkx as nx
import matplotlib.pyplot as plt
from swimlane_tools import SwimlaneDocumentation, SwimlaneBuildTool
```

```python
notebook_name = 'Swimlane Tools Examples.ipynb'
build_file_name_1 = 'Build Test 1'
build_file_name_2 = 'Build Test 2'
build_file_name_3 = 'Build Test 3'
build_file_name_4 = 'Build Test 4'

sd1 = SwimlaneDocumentation(build_file_name_1, notebook_name, used_by=['Build_Test_4'])
sd1.add_node('BT1 OUTPUT')
sd1.add_node('input1.csv', used_by=['BT1 OUTPUT'])
sd1.add_node('country codes', used_by=['BT1 OUTPUT'])

sd2 = SwimlaneDocumentation(build_file_name_2, notebook_name, used_by=['Build_Test_4'])
sd2.add_node('BT2 OUTPUT')
sd2.add_node('input2.csv', used_by=['BT2 OUTPUT'])

sd3 = SwimlaneDocumentation(build_file_name_3, notebook_name, used_by=['Build_Test_4'])
sd3.add_node('BT3 OUTPUT')
sd3.add_node('input3.csv', used_by=['BT3 OUTPUT'])
sd3.add_node('country codes', used_by=['BT3 OUTPUT'])

sd4 = SwimlaneDocumentation(build_file_name_4, notebook_name, used_by=['Final Vis'])
sd4.add_node('Final Vis')
sd4.add_node('final output.csv', used_by='Final Vis')
sd4.add_node('BT1 OUTPUT', used_by=['final output.csv'])
sd4.add_node('BT2 OUTPUT', used_by=['final output.csv'])
sd4.add_node('BT3 OUTPUT', used_by=['final output.csv'])
```

### BuildTool Examples:

```python
bt1 = SwimlaneBuildTool()
```

```python
standard_graph = bt1.get_initialized_DiGraph()

# Plot standard visualization
plt.figure(figsize=(10, 6))
pos = bt1.pos  # Assuming this is your standard positions data  node_color=[sbt.node_colors[node]
nx.draw(standard_graph, pos=pos, node_color=[bt1.node_colors[node] for node in standard_graph.nodes], 
        with_labels=True, node_size=1000, font_size=15, arrows=True)
plt.title('Standard Visualization')
plt.show()
```

```python
standard_graph = bt1.get_initialized_DiGraph()

# Plot standard visualization
plt.figure(figsize=(10, 6))
pos = bt1.pos  # Assuming this is your standard positions data  node_color=[sbt.node_colors[node]
nx.draw(standard_graph, pos=pos, node_color=[bt1.node_colors[node] for node in standard_graph.nodes], 
        node_shape='o', font_size=8, font_color='black', font_weight='bold', arrowsize=20, connectionstyle='arc3, rad=0.1')
nx.draw_networkx_labels(standard_graph, bt1.label_pos, font_size=8, font_color='black', font_weight='bold')

plt.title('Arc Visualization')
plt.show()
```

```python
notebook_graph = bt1.get_initialized_notebook_DiGraph()

# Plot notebook visualization  
plt.figure(figsize=(10, 6))
notebook_pos = bt1.notebook_pos  # Assuming this is your notebook positions data
nx.draw(notebook_graph, pos=notebook_pos, node_color=[bt1.notebook_node_colors[node] for node in notebook_graph.nodes], 
        with_labels=True, node_size=1000, font_size=15, arrows=True)
plt.title('Notebook Visualization')
plt.show()
```
